ลบแถวทั้งหมดถ้าคอลัมน์ใดคอลัมน์หนึ่งที่มีค่า NaN (ค่าหาย) ออกจาก DataFrame

หาค่าเฉลี่ยของอายุ และ แทนที่ค่าหายของอายุด้วยค่าเฉลี่ยของอายุ

```python
mean_age = df['อายุ'].mean()
print(f"ค่าเฉลี่ยอายุ = {mean_age:.6f} ปี")
```

```python
df['อายุ'].fillna(mean_age, inplace=True)
df
```

หาค่าคณะที่พบบ่อยที่สุด และเติมค่าหายของคณะด้วยคณะที่พบมากที่สุด

```python
most_faculty = df['คณะ'].mode()[0]
print(f"คณะที่พบบ่อยที่สุด = {most_faculty}")
```

```python
df['คณะ'].fillna(most_faculty, inplace=True)
df
```

ทำข้อมูลของชื่อคณะให้เป็นมาตราฐานเพื่อลดความซ้ำซ้อนของข้อมูลจากการพิมพ์ไม่เหมือนกัน

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์', 'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรม', 'วิศวกรรมศาสตร์', 'วิศวฯ', 'Engineering', 'วิศวะ']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
df
```

หาค่าที่อยู่นอกช่วงปกติและตัดค่าที่อยู่นอกช่วงปกติออก ช่วยทำให้ข้อมูลไม่ถูกค่าผิดปกติรบกวน

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
df_no_outliers
```

แปลงข้อมูลของคณะมาตราฐานให้เป็น One-Hot Encoding เพื่อให้ AI สามารถเข้าใจได้

```python
df_encoded = pd.get_dummies(df, columns=['คณะ_มาตรฐาน'])
df_encoded
```

แปลงเวลาเป็น Datetime เพื่อดึงข้อมูลชั่วโมง และใช้ pd.cut จัดกลุ่มช่วงเวลา (สร้างข้อมูลใหม่เนื่องจากข้อมูลเดิมไม่มีช่วงเวลาให้จัดกลุ่ม)

```python
coffee_data = pd.DataFrame({
    'วันเวลา': pd.to_datetime(['2024-01-15 07:30', '2024-01-15 12:00',
                               '2024-01-15 16:30', '2024-01-16 08:00']),
    'ยอดขาย': [1500, 3200, 2100, 1800]
})

coffee_data['ชั่วโมง'] = coffee_data['วันเวลา'].dt.hour
coffee_data['วันในสัปดาห์'] = coffee_data['วันเวลา'].dt.day_name()
coffee_data['ช่วงเวลา'] = pd.cut(coffee_data['ชั่วโมง'],
                                  bins=[0, 10, 14, 18, 24],
                                  labels=['เช้า', 'กลางวัน', 'บ่าย', 'เย็น'])

coffee_data[['วันเวลา', 'ช่วงเวลา', 'ยอดขาย']]
```

รวมข้อมูลส่วนสูงและอายุเพื่อสร้างค่า BMI

```python
df['BMI'] = df['น้ำหนัก_kg'] / (df['ส่วนสูง_cm']/100)**2
df['สุขภาพ'] = pd.cut(df['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])

df[['ชื่อ', 'BMI', 'สุขภาพ']]
```
